```python
def load_ingredients():

    food_dict = load_dictionary()
    ingredients = []
    all_meals = list(food_dict.keys())

    for i in range(len(food_dict)):
        selected = food_dict[all_meals[i]]['ingredients'] + food_dict[all_meals[i]]['optional']
        for j in range(len(selected)):
            if selected[j] not in ingredients:
                ingredients.append(selected[j])


    ingredients.append('f')
    
    return ingredients

ingredients_list = load_ingredients()
for greg in ingredients_list:
    print(greg, end = ', ')
```

```python
#DICTIONARY FINDER EXPLANATION

#1 Variable Setup, possible meals list, possible meals dictionary, list of all meals

#2 Loops through list of all meal names, sets up variables for the current meals list of ingredients and its name e.g c_l_name =
#'carbonara' c_l = ['pasta','cheese']. Sets up a new loop to go through all of the users ingredients, and for each iteration
# sets a target. If the users ingredient (target) is in the ingredients list of the meal, it is appended to the possible meals
# list. This initial p_meals_list has duplicates, but a new variable is set up to have that list but with duplicates removed.

#3 So now we have a list of all the meals that the user matched at least one ingredient with, and the same list but with
# duplicate possible meals removed. Now for the tricky/annoying part. loop through the non duplicate possible meals list. Big
# variable set up here. y and b are lists that are just placeholders and used later for the needs and has part. The current 
# meal selected in the loop is given a variable to make life easier. Then the possible meals dictionary is set up with 4 lists
# named has, needs, optional has and optional needs with pretty self explanatory reasons. needs and optional needs are set up
# to just be ingredients and optional of their respective meal from the meals dictionary. Almost there... Now we loop through
# the users ingredients. If the users-ingredient appears in 'needs', it is moved to 'has' using computer science magic
# (making needs list = x, removing currently selected user_ingredient, making 'needs' = x, then appending currently selected
# user ingredient to y and making 'has' = y). The same is done for optional and finally matches is just the length of has as
# as that represents how many user ingredients matched the needs. Then badabing badaboom you have successfully scanned a
# dictionary in python. Now all you need to do is print it out nicely and thats basically the heart of the program


def dictionary_finder(user_ingreds,meals):
    
    #1
    p_meals_list,p_meals_dict,all_meals_list = [], {}, list(meals.keys())
    
    #2
    for i in range(len(all_meals_list)):
        c_l, c_l_name = meals[all_meals_list[i]]['ingredients'], all_meals_list[i]           
        for j in range(len(user_ingreds)):
            target = user_ingreds[j]
            if target in c_l:
                p_meals_list.append(c_l_name)
    no_dupe_meals = duplicate(p_meals_list)
    
#    print('no dupe meals: ',no_dupe_meals)
#    print('meals with duplicates: ',meals)
    
    #3
    for i in range(len(no_dupe_meals)):
        #(YIPEE!)
        y = []
        b = []
        current_meal = no_dupe_meals[i]
        p_meals_dict[current_meal] = {}
        p_meals_dict[current_meal]['needs'] = meals[current_meal]['ingredients']
        p_meals_dict[current_meal]['optional_needs'] = meals[current_meal]['optional']
        p_meals_dict[current_meal]['has'] = []
        p_meals_dict[current_meal]['optional_has'] = []
        
        for j in range(len(user_ingreds)):
            target = user_ingreds[j]
            if target in p_meals_dict[current_meal]['needs']:
                #I HAVE A FEELING THIS COULD BE SIMPLIFIED \/\/\/\/\/\/\
                x = p_meals_dict[current_meal]['needs']
                x.remove(target)
                p_meals_dict[current_meal]['needs'] = x
                y.append(target)
                p_meals_dict[current_meal]['has'] = y   
            elif target in p_meals_dict[current_meal]['optional_needs']:
                a = p_meals_dict[current_meal]['optional_needs']
                a.remove(target)
                p_meals_dict[current_meal]['optional_needs'] = a
                b.append(target)
                p_meals_dict[current_meal]['optional_has'] = b 
        p_meals_dict[current_meal]['matches'] = len(p_meals_dict[no_dupe_meals[i]]['has'])
    return p_meals_dict
```

```python
def duplicate(l):
    wordList = []
    badList = []
    for element in l:
        if element in wordList:
            badList.append(element)
        else:
            wordList.append(element)    
    return wordList
```

```python
def output_print(p,meals_dictionary):
    keys = list(p.keys())
    c_unc = False
    
    for i in range(len(p)):
        
        if meals_dictionary[keys[i]]['cooked?'] != c_unc:
            c_unc = meals_dictionary[keys[i]]['cooked?']
        if c_unc == True:
            c_unc_str = 'cooked'
        else:
            c_unc_str = 'uncooked'
        
        print('>'+(keys[i]).title(),'\n=====================\nMeal matched',len(p[keys[i]]['has']),'inventory ingredients.')
        print('Meal is',c_unc_str+', commonly eaten at',meals_dictionary[keys[i]]['tod?']+'.\n')
        print('Necessary Ingredients in Inventory:')
        for j in range(len(p[keys[i]]['has'])):
            print(('-'+p[keys[i]]['has'][j]).title())
        print('')
        print('Necessary Ingredients NOT in Inventory:')
        
        if len(p[keys[i]]['needs']) == 0:
            print('You have everything!')
        else:
            for k in range(len(p[keys[i]]['needs'])):
                print(('-'+p[keys[i]]['needs'][k]).title())
        if len(p[keys[i]]['optional_has']) !=0:
            print('')
            print('Optional Ingredients in Inventory:')

            for g in range(len(p[keys[i]]['optional_has'])):
                print(('-'+p[keys[i]]['optional_has'][g]).title())
        
        print('')
        print('Optional Ingredients NOT in Inventory:')
        
        if len(p[keys[i]]['optional_needs']) == 0:
            print('Nothing! You have everything!')
        else:
            for k in range(len(p[keys[i]]['optional_needs'])):
                print(('-'+p[keys[i]]['optional_needs'][k]).title())
        print('')
```
